## Step 4: Data Arrangement

```python
irisf.drop('Id',axis=1,inplace=True)
```

```python
sp = {'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3}
```

```python
irisf.Species = [sp[i] for i in irisf.Species]
```

```python
irisf
```

```python
X = irisf.iloc[:, 0:4]
```

```python
X
```

```python
Y = irisf.iloc[:,4]
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

## Step 5: Training The Model

```python
irismodel = LinearRegression()
```

```python
irismodel.fit(X,Y)
```

```python
irismodel.score(X,Y)
```

```python
irismodel.coef_
```

```python
irismodel.intercept_
```

```python
#Applying KNN algorithm to predivt the accuracy
from sklearn.metrics import accuracy_score
irismodel = KNeighborsClassifier()
irismodel.fit(X_train,Y_train)
predictions = irismodel.predict(X_test)
print("accuracy score = ",accuracy_score(Y_test, predictions))
```

```python
#Support Vector Machine ModelTest
from sklearn.svm import SVC
irismodel = SVC()
irismodel.fit(X_train,Y_train)
predictions = irismodel.predict(X_test)
print("accuracy score = ",accuracy_score(Y_test, predictions))
```

```python
#With Random Forest (Decision Tree Classification)
from sklearn.ensemble import RandomForestClassifier
irismodel = RandomForestClassifier(n_estimators=5)
irismodel.fit(X_train,Y_train)
predictions = irismodel.predict(X_test)
print("accuracy score = ",accuracy_score(Y_test, predictions))
```

## Step 6: Making Predictions

```python
Y_prediction = irismodel.predict(X_test)
```

## Step 7: Mean Squared Error

```python
#Evolution Of Iris Flower Model
print("Mean squared error: %.2f" % np.mean((Y_prediction - Y_test) ** 2))
```
